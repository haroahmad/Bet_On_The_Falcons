# Logistic Regression
<br>

* Predicts which class or category an observation falls under 
<br>
<br>
* Uses multiple independent features to predict on a dependent variable
<br>
<br>
* Provides a probability score for the event (likelihood of winning)

# Advantages
<br>

* Easy to understand and interpret (think Linear Regression)
<br>
<br>
* Requires minimal computation resources
<br>
<br>
* Provides feature importance and assessment

# Disadvantages
<br>

* Smaller datasets are prone to overfitting
<br>
<br>
* Requires scaled and/or normalized data
<br>
<br>
* Sensitive to multicollinearity

# Random Forest
<br>

* Built from decision trees (see below)

<br>
<br>
* Combines the results of many decision trees to form a more accurate prediction


# Advantages
<br>

* Accuracy score is generally high due to "wisdom of the crowd"
<br>
<br>
* Provides feature importance for interpretability
<br>
<br>
* Overfitting is reduced given enough trees are used

# Disadvantages
<br>

* More trees = more resources. Can be slow for real-time predictions
<br>
<br>
* Difficult to interpret when dealing with hundreds of trees: 
<br>
<br>


# XGBoost (Extreme Gradient Boosting)
<br>

* In short, Decision Trees on steroids
<br>
<br>
* Combines a Decision Tree with Gradient Boosting (aggregates multiple predictions)
<br>
<br>
* Popular! Utilized in many winning Kaggle competitions

# Advantages
<br>

* Many times faster than other models, and better performing
<br>
<br>
* Built-in features that prevent overfitting (L1 and L2 regularization)
<br>
<br>
* Works with most datasets

# Disadvantages
<br>

* Categorical features must be label encoded before processing
<br>
<br>
* Similar to Random Forest, can be difficult to interpret (black-box)
<br>
<br>
* Sensitive to outliers
